```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
sigmoid = np.vectorize(lambda x: 1.0/(1.0 + np.exp(-x)))
```

An example of GAM GLM where a single variable x is considered and a BSplines smoother function is used.

```python
# GAM GLM
from statsmodels.sandbox.gam_gsoc2015.smooth_basis import BSplines
from statsmodels.sandbox.gam_gsoc2015.gam import GLMGam

n = 200
x = np.linspace(-10, 10, n)
y = x * x + np.random.normal(0, 5, n)
y -= y.mean()

x = x - x.mean()

bsplines = BSplines(x, degree=[5], df=[10])

alphas = [0, 0.001, 0.06, 1]
for i, alpha in enumerate(alphas):
    plt.subplot(2, 2, i+1)

    # train the model
    
    glm_gam = GLMGam(y, bsplines, alpha=alpha)
    res_glm_gam = glm_gam.fit(method='nm', max_start_irls=0,
                              disp=0, maxiter=5000, maxfun=5000)
    plt.plot(x, np.dot(bsplines.basis_, res_glm_gam.params))
    plt.plot(x, y, '.')
    plt.title('GLM alpha=' + str(alpha))
plt.tight_layout()
plt.show()
```

GLM GAM with binomial output and 2 input variables

```python
from statsmodels.sandbox.gam_gsoc2015.gam import LogitGam

np.random.seed(0)
n = 100
x1 = np.linspace(-1, 1, n)
x2 = np.linspace(0, 1, n)

poly = x1*x1*x1 + x2*x2 + np.random.normal(0, .5, n)
y = sigmoid(poly)
mu = y.mean()
yc = y.copy()
y[y > mu] = 1
y[y <= mu] = 0

degree1 = 3
degree2 = 3
df = 3

x = np.vstack([x1, x2]).T
bsplines = BSplines(x, [df, df], [degree1, degree2])

alpha = [30] * 2

mLG = LogitGam(y, bsplines, alpha=alpha)
res_mLG = mLG.fit(maxiter=1000, tol=1e-13, method='nm')

y_est = np.dot(bsplines.basis_, res_mLG.params)
y_est = sigmoid(y_est)

plt.plot(y_est, label='y est')
plt.plot(y, '.', label='y')
plt.legend(loc='best')
plt.ylim(-0.2, 1.2)
plt.show()
```

Example of the Pirls algorithm

```python
from statsmodels.sandbox.gam_gsoc2015.smooth_basis import CubicSplines

n = 500
x = np.random.uniform(-1, 1, n)

y = 10*x**3 - 10*x + np.random.normal(0, 1, n)

y -= y.mean()
cs = CubicSplines(x, [10])

# required only to initialize the gam. they have no influence on the result.
for i, alpha in enumerate([0, 1, 5, 10]):

    gam = GLMGam(y, cs, alpha=alpha)
    gam_res = gam.fit()
    y_est = gam_res.predict(cs.basis_)

    plt.subplot(2, 2, i+1)
    plt.plot(x, y, '.')
    plt.plot(x, y_est, '.')
    plt.title('alpha=' + str(alpha))

plt.tight_layout()
plt.show()
```

Example of the multivariate pirls algorithm


```python
np.random.seed(0)
# Multivariate GAM PIRLS
n = 500
x1 = np.random.uniform(-1, 1, n)
y1 = 10*x1**3 - 10*x1 + np.random.normal(0, 3, n)
y1 -= y1.mean()
x2 = np.random.uniform(-1, 1, n)
y2 = x2**2 + np.random.normal(0, 0.5, n)
y2 -= y2.mean()

y = y1 + y2

x = np.zeros(shape=(n, 2))
x[:, 0] = x1
x[:, 1] = x2

cs = CubicSplines(x, df=[10, 10])

i = 0
for alpha1, alpha2 in zip([0, 1, 2], [0, 1, 2]):
    
    gam = GLMGam(y, cs, alpha=[alpha1, alpha2])
    gam_results = gam.fit()
    y1_est = np.dot(cs.basis_[:, cs.mask[0]], gam_results.params[cs.mask[0]])
    y2_est = np.dot(cs.basis_[:, cs.mask[1]], gam_results.params[cs.mask[1]])

    y1_est -= y1_est.mean() # TODO: the estimate is good but has a very large mean. Why is this happening?
    y2_est -= y2_est.mean()

    i += 1
    plt.subplot(3, 2, i)
    plt.title('x1  alpha=' + str(alpha1))
    plt.plot(x1, y1, '.', label='Real', c='green')
    plt.plot(x1, y1_est, '.', label='Estimated', c='blue')

    i += 1
    plt.subplot(3, 2, i)
    plt.title('x2  alpha=' + str(alpha2))
    plt.plot(x2, y2, '.', label='Real', c='green')
    plt.plot(x2, y2_est, '.', label='Estimated', c='blue')

plt.tight_layout()
plt.show()
```

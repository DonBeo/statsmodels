```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.sandbox.gam_gsoc2015.smooth_basis import (UnivariateBSplines, UnivariatePolynomialSmoother, BSplines,
                                                           PolynomialSmoother, CubicSplines)
from statsmodels.sandbox.gam_gsoc2015.gam import UnivariateGamPenalty, LogitGam, GLMGam, MultivariateGamPenalty
import statsmodels.api as sm
from statsmodels.sandbox.gam_gsoc2015.gam_cross_validation.gam_cross_validation import MultivariateGAMCV, MultivariateGAMCVPath
from statsmodels.sandbox.gam_gsoc2015.gam_cross_validation.cross_validators import KFold

%matplotlib inline
sigmoid = np.vectorize(lambda x: 1.0/(1.0 + np.exp(-x)))
```

An example of GAM GLM where a single variable x is considered and a BSplines smoother function is used.

```python
# GAM GLM
n = 200
x = np.linspace(-10, 10, n)
y = x * x + np.random.normal(0, 5, n)
y -= y.mean()

x = x - x.mean()

bsplines = BSplines(x, degrees=[5], dfs=[10])

alphas = [0, 0.001, 0.06, 1]
for i, alpha in enumerate(alphas):
    plt.subplot(2, 2, i+1)

    # train the model
    gp = MultivariateGamPenalty(bsplines, alphas=[alpha])
    glm_gam = GLMGam(y, bsplines.basis_, penal = gp)
    res_glm_gam = glm_gam.fit(method='nm', max_start_irls=0,
                              disp=0, maxiter=5000, maxfun=5000)
    plt.plot(x, np.dot(bsplines.basis_, res_glm_gam.params))
    plt.plot(x, y, '.')
    plt.title('GLM alpha=' + str(alpha))
plt.tight_layout()
plt.show()
```

GLM GAM with binomial output and 2 input variables

```python
np.random.seed(0)
n = 100
x1 = np.linspace(-1, 1, n)
x2 = np.linspace(0, 1, n)

poly = x1*x1*x1 + x2*x2 + np.random.normal(0, .5, n)
y = sigmoid(poly)
mu = y.mean()
yc = y.copy()
y[y > mu] = 1
y[y <= mu] = 0

degree1 = 3
degree2 = 3
df = 3

x = np.vstack([x1, x2]).T
bsplines = BSplines(x, [df, df], [degree1, degree2])

alpha = [30] * 2
mgp = MultivariateGamPenalty(bsplines, alphas=alpha, wts=[1, 1])

mLG = LogitGam(y, bsplines.basis_, penal=mgp)
res_mLG = mLG.fit(maxiter=1000, tol=1e-13, method='nm')

y_est = np.dot(bsplines.basis_, res_mLG.params)
y_est = sigmoid(y_est)

plt.plot(y_est, label='y est')
plt.plot(y, '.', label='y')
plt.legend(loc='best')
plt.ylim(-0.2, 1.2)
plt.show()
```

Example of the Pirls algorithm

```python
n = 500
x = np.random.uniform(-1, 1, n)

y = 10*x**3 - 10*x + np.random.normal(0, 1, n)

y -= y.mean()
cs = CubicSplines(x, 10)

# required only to initialize the gam. they have no influence on the result.
dummy_smoother = PolynomialSmoother(x, [2])
gp = MultivariateGamPenalty(dummy_smoother, alphas=[0])
for i, alpha in enumerate([0, 1, 5, 10]):

    gam = GLMGam(y, cs.basis_, penal=gp)
    gam_res = gam._fit_pirls(y=y, spl_x=cs.basis_, spl_s=cs.s, alpha=alpha)
    y_est = gam_res.predict(cs.basis_)

    plt.subplot(2, 2, i+1)
    plt.plot(x, y, '.')
    plt.plot(x, y_est, '.')
    plt.title('alpha=' + str(alpha))

plt.tight_layout()
plt.show()
```

Example of the multivariate pirls algorithm


```python
np.random.seed(0)
# Multivariate GAM PIRLS
n = 500
x1 = np.random.uniform(-1, 1, n)
y1 = 10*x1**3 - 10*x1 + np.random.normal(0, 3, n)
y1 -= y1.mean()
x2 = np.random.uniform(-1, 1, n)
y2 = x2**2 + np.random.normal(0, 0.5, n)
y2 -= y2.mean()

Y = y1 + y2
cs1 = CubicSplines(x1, 10)
cs2 = CubicSplines(x2, 10)

spl_X = np.hstack([cs1.basis_, cs2.basis_])
spl_S = [cs1.s, cs2.s]

n_var1 = cs1.basis_.shape[1]
n_var2 = cs2.basis_.shape[1]

dummy_smoother = BSplines(x1, dfs=[10], degrees=[3])
gp = MultivariateGamPenalty(dummy_smoother, alphas=[1])
gam = GLMGam(y1, cs1.basis_, penal=gp)

i = 0
for alpha1, alpha2 in zip([0, 1, 2], [0, 1, 2]):

    gam_results = gam._fit_pirls(Y, spl_X, spl_S, alpha=[alpha1, alpha2])
    y1_est = np.dot(cs1.basis_, gam_results.params[:n_var1])
    y2_est = np.dot(cs2.basis_, gam_results.params[n_var1:])

    y1_est -= y1_est.mean() # TODO: the estimate is good but has a very large mean. Why is this happening?
    y2_est -= y2_est.mean()

    i += 1
    plt.subplot(3, 2, i)
    plt.title('x1  alpha=' + str(alpha1))
    plt.plot(x1, y1, '.', label='Real', c='green')
    plt.plot(x1, y1_est, '.', label='Estimated', c='blue')

    i += 1
    plt.subplot(3, 2, i)
    plt.title('x2  alpha=' + str(alpha2))
    plt.plot(x2, y2, '.', label='Real', c='green')
    plt.plot(x2, y2_est, '.', label='Estimated', c='blue')

plt.tight_layout()
plt.show()
```

CV Example


```python
np.random.seed(0)


def sample_metric(y1, y2):

    return np.linalg.norm(y1 - y2)/len(y1)

n = 300
df = [10]
degree = [6]

x = np.random.uniform(-1, 1, n)
y = np.sin(x) / x + np.random.normal(0, .1, n)

bsplines = BSplines(x, degrees=degree, dfs=df)

gam = GLMGam
alphas = [np.linspace(0, 2, 10)]
k = 3
cv = KFold(k_folds=k, shuffle=True)

gam_cv = MultivariateGAMCVPath(smoothers=bsplines, alphas=alphas, gam=gam, 
                               cost=sample_metric, y=y, cv=cv)
gam_cv_res = gam_cv.fit(method='nm', max_start_irls=0,
                        disp=0, maxiter=10000, maxfun=10000)

print('alpha cv =', gam_cv.alpha_cv_)

gp = MultivariateGamPenalty(bsplines, alphas=gam_cv.alpha_cv_)
glm_gam = GLMGam(y, bsplines.basis_, penal=gp)
res_glm_gam = glm_gam.fit(method='nm', max_start_irls=0,
                          disp=0, maxiter=10000, maxfun=10000)
y_est = res_glm_gam.predict(bsplines.basis_)

plt.plot(x, y, '.', label='y')
plt.plot(x, y_est, '.', label='y est')
plt.legend()
plt.show()
```
